```python
_y.std()
```

# Question 2:
Sum together every number from 0 to 10000 except for those than can be divided by 4 or 7. Do this in one line of code

```python
# Using vectorize

def check(x):
    if(x%4 == 0 or x%7==0):
        return False
    else:
        return True
x = np.arange(0, 10001, 1)

np.sum(x[np.vectorize(check)(x)])
```

```python
# Oneliner

np.sum(x[(x%4!=0)*(x%7!=0)])
```

## Linear system

$\left\{ 
\begin{array}{c}
x+y+3z+t=0 \\ 
x-y-z-t=0 \\ 
3x+y+5z+3t=0 \\
x+5y+11z+8t=0
\end{array}
\right.
$

```python
A = np.array(
    [
        [1,1,3,1],
        [1,-1,-1,-1],
        [3,10,5,3],
        [1,5,11,8]
    ]
)
```

```python
A
```

```python
A0 = np.zeros(4)
A0
```

```python
np.linalg.solve(A,A0)
```
