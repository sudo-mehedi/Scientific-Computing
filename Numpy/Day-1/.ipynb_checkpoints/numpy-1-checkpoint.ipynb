### Integrals and derivative

```python
x = np.linspace(1, 10, 100)
y  = 1/x**2 * np.sin(x)

dydx = np.gradient(y,x)


y_int = np.cumsum(y)*(x[1]-x[0])
```

```python
plt.plot(x,y)
plt.plot(x, dydx)
plt.plot(x, y_int)
```

 ## Question 1:
Let $y=e^{-x/10} \\sin(x)$. Consider 10000 $x$ intervals in the range $[0,10]$
1. Plot the function $y$ vs. $x$ in the range $[0, 10]$. -done
2. Compute the mean and standard deviation of $y$ for $x$ values in $[4,7]$
3. For $x$ in the range $[4,7]$, find the value $y_m$ such that 80% of $y$ values are less than $y_m$
4. Plot $dy/dx$ vs $x$
5. Find the locations where $dy/dx=0$

```python
def f(x):
    return np.exp(-x/10) * np.sin(x)
```

```python
x = np.linspace(0, 10, 10000)
y = f(x)
```

```python
plt.plot(x, y)
```

```python
def check(x):
    if ( x>4 and x < 7):
        return True
    else:
        return False

_y = y[np.vectorize(check)(x) == True]
_y.mean()
```

```python
_y.std()
```

```python
np.percentile(_y, 80)
```

```python
dydx = np.gradient(y,x)
```

```python
plt.plot(x,y)
plt.plot(x, dydx)
```
